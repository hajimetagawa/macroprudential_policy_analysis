```python
import pandas as pd

urls = ["https://stats.bis.org/api/v2/data/dataflow/BIS/WS_TC/2.0/?format=csv"]

df = pd.concat([pd.read_csv(url) for url in urls])
df.head()
```

```python
# 'UNIT_TYPE' が '770' かつ'TC_BORROWERS' が 'C'の行だけを抽出
df_filtered = df[
    (df['UNIT_TYPE'] == '770') &
    (df['TC_BORROWERS'] == 'C')
].copy()

# 必要な列だけを抽出
df_filtered = df_filtered[[
    'BORROWERS_CTY',
    'TIME_PERIOD',
    'OBS_VALUE'
]].copy()

# 結果を確認
df_filtered.head()
```

```python
# ファイルパスの指定（1階層上のフォルダに保存）
output_path = '../data/bis_CREDIT TO NON FINANCIAL SECTOR.csv'

# CSVとして保存（UTF-8でBOMなし、インデックス列なし）
df_filtered.to_csv(output_path, index=False)
```
