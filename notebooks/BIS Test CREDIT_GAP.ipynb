```python
import pandas as pd

urls = ["https://stats.bis.org/api/v2/data/dataflow/BIS/WS_CREDIT_GAP/1.0/?format=csv"]

df = pd.concat([pd.read_csv(url) for url in urls])
```

```python
df.head()
```

```python
# 'CG_DTYPE' が 'B' の行だけを抽出
df_filtered = df[df['CG_DTYPE'] == 'B'].copy()

# 必要な列だけを抽出
df_filtered = df_filtered[[
    'BORROWERS_CTY',
    'TIME_PERIOD',
    'OBS_VALUE'
]].copy()

# 結果を確認
df_filtered.head()
```

```python
# ファイルパスの指定（1階層上のフォルダに保存）
output_path = '../data/bis_data_CREDIT GAP.csv'

# CSVとして保存（UTF-8でBOMなし、インデックス列なし）
df_filtered.to_csv(output_path, index=False)
```
